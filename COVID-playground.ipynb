```python
import pandas as pd
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.cdc.gov", None)

# First 5000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("n8mc-b4w4", limit=5000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
results_df
```

```python
results_df = results_df.drop(['case_onset_interval'], axis=1)
results_df = results_df.drop(['current_status'], axis=1)
results_df = results_df.drop(['case_positive_specimen'], axis=1)
results_df = results_df.drop(['symptom_status'], axis=1)
results_df = results_df.drop(['state_fips_code'], axis=1)
results_df = results_df.drop(['county_fips_code'], axis=1)
results_df
```
