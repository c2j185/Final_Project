# DataViz Data Cleanup

```python
import pandas as pd
```

```python
file = "DataVisCleanDataLKW.csv"
full_results_df1 = pd.read_csv(file)
full_results_df1
```

```python
# Make a copy of the df with the empty column "outcome"
results_df = full_results_df1.copy()

results_df["outcome"]=""
results_df
```

```python
# Populate "outcome" column
for idx, row in results_df.iterrows():
    if row['death'] == "Yes":
        row['outcome'] = "severe"
 
    elif row['hospitalization'] != "Yes" and row['icu'] != "Yes" and row['death'] != "Yes":
        row['outcome'] = "mild"
           
    elif row['hospitalization'] != "Yes" and row['icu'] == "Yes" and row['death'] != "Yes":
        row['outcome'] = "severe"
        
    elif row['hospitalization'] == "Yes" and row['icu'] != "Yes" and row['death'] != "Yes":
        row['outcome'] = "severe"

    elif row['hospitalization'] == "Yes" and row['icu'] == "Yes" and row['death'] != "Yes":
        row['outcome'] = "severe"
    
    else:
        row['outcome'] = "mild"


# Print the contents and counts of the "outcome" column
outcome_counts = results_df.outcome.value_counts()
print(outcome_counts)
```

```python
results_df = results_df.drop(['hospitalization'], axis=1)
results_df = results_df.drop(['icu'], axis=1)
results_df = results_df.drop(['death'], axis=1)
results_df
```
